## An example of arithmetic operations on ndarrays 
## *Un ejemplo de operaciones aritméticas entre ndarrays*

### Images
Using images is a nice way to understand what's happening. That's why I'm going to use Matplotlib; if I would like to work only on arrays with numbers, it's more likely to use NumPy library instead.

### *Imágenes*
*Usar imágenes es una buena forma de entender qué está sucediendo. Por eso, voy a usar Matplotlib; si solo fuese a trabajar con matrices de números, probablemente usaría la librería NumPy nada más.*

```python
import matplotlib.pyplot as plt
image1 = plt.imread("C:/Users/Mike/Pictures/minyb.jpg") # <-- file from which I would get the first group of three arrays. 
                                                        # de este archivo obtendré el primer grupo de tres matrices.
print("Flags:", image1.flags)
print("Shape:", image1.shape)  # <-- I will be able to add two arrays only if they have the same lines/columns shape.
                               # solo podré sumar dos matrices que tengan la misma cantidad de líneas/columnas.
print("Type:", type(image1))   # <-- To show that I'm working on arrays. // Muestra que estoy trabajando en arrays.
plt.figure
plt.imshow(image1)
```

**The following // A continuación.**

(EN) The same previous commands, but these are for the second group of arrays, which will be added to the first one.

(ES) Los mismos comandos usados previamente, pero estos son para el segundo grupo de matrices, estas serán sumadas a las primeras.

```python
image2 = plt.imread("C:/Users/Mike/Pictures/minp.jpg")
print("Flags:", image2.flags)
print("Shape:", image2.shape)
print("Type:", type(image2))
plt.figure
plt.imshow(image2)
```

### Adding ndarrays. // Suma de ndarrays.
(EN) The following first line of code does the addition. If I had import the NumPy library (import numpy), I will be able to do the same using:

added_images = numpy.add(image1, image2)

(ES) La primera línea de código a continuación realiza la suma. Si hubiese importado la librería NumPy (import numpy), también podría hacer la suma usando:

added_images = numpy.add(image1, image2)

```python
added_images = image1 + image2
print("Shape:", added_images.shape)
print("Type:", type(added_images))
plt.figure
plt.imshow(added_images)
```

### Notice that: // Obsérvese que:
*(EN)*
* Both arrays (or groups of arrays) have to have exactly the same shape.
* Since I'm working on images, what I'm adding are the values (colors) of each pixel from the first image and the second one. That's why is possible to see some pixels which seem to have "wrong" colors (like the shoes of the leftside character, or the mouth of the rightside one).

    It looks like errors, but actually that's right. Because those pixels, in their original images, have values that produce that result once their are added.
    
    For those of you who want a deeper explanation, I will leave it in the following section.

*(ES)*
* Ambas matrices (o grupos de matrices) tienen la misma forma exactamente.
* Como estoy trabajando con imágenes, lo que estoy sumando son los valores (colores) de cada pixel de la primera imagen y la segunda. Es por eso que es posible ver algunos pixeles que parecen tener colores "erróneos" (como los zapatos del personaje que está a la izquierda, o la boca del que está a la derecha).

    Parecen errores pero, en realidad, es lo correcto. Porque esos pixeles, en sus imágenes originales, tienen valores que producen ese resultado al ser sumados.
    
    Para aquellos de ustedes que deseen una explicación más profunda, la dejaré a continuación.

### A little adjustment // Un pequeño ajuste.
**(EN)** Notice that the absolute absence of color has zero as its value (black), and the absolute presence of color has 255 as its value (white).

   That's why, in the following lines of code, I'm setting a range of values to be -255 in both images, doing this I ensure to have zero or underzero values (which means black color pixels) after the addition, and that's what I need to have a right view of the shoes or the mouth.
    
   I'm leaving the area where the hand and the hair adds to each other "without modification" to keep the esence of this article.
    
**(ES)** Téngase en cuenta que la ausencia absoluta de color tiene por valor el cero (negro), y la presencia absoluta de color tiene por valor 255 (blanco).

   Por eso, en las siguientes líneas de código, selecciono un rango de valores y les asigno el valor -255, de esta forma aseguro que el valor después de sumar las matrices será cero o inferior a cero (lo que significa que será un pixel negro), y eso es lo que necesito para que los zapatos o la boca se vean de una forma correcta.
    
   Estoy dejando la sección donde se suma la mano de uno y el cabello del otro "sin modificar" para mantener la esencia del artículo.

```python
image1.setflags(write=True)
cleaning_bright = image1 < 20
image1[cleaning_bright] = -255
plt.imshow(image1)
```

```python
image2.setflags(write=True)
cleaning_bright = image2 < 20
image2[cleaning_bright] = -255
plt.imshow(image2)
```

```python
added_images = image1 + image2
print("Shape:", added_images.shape)
print("Type:", type(added_images))
plt.figure
plt.imshow(added_images)
```
